## Standardizing Independent Data using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x1 = sc.fit_transform(x)
```

```python
x1
```

## Training & Testing Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1, y, train_size=0.75,random_state=101)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(x_train, y_train)
```

```python
y_pred_test = logit.predict(x_test)
y_pred_train = logit.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score,precision_score,recall_score,roc_auc_score,roc_curve
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(confusion_matrix(y_train, y_pred_train)), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix for training data', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(confusion_matrix(y_test, y_pred_test)), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix for testing data', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print("Accuracy of training dataset:",accuracy_score(y_train, y_pred_train)*100,'%')
print("Precision of training dataset:",precision_score(y_train, y_pred_train)*100,'%')
print("Recall of training dataset:",recall_score(y_train, y_pred_train)*100,'%')
```

```python
print("Accuracy of testing dataset:",accuracy_score(y_test, y_pred_test)*100,'%')
print("Precision of testing dataset:",precision_score(y_test, y_pred_test)*100,'%')
print("Recall of testing dataset:",recall_score(y_test, y_pred_test)*100,'%')
```

```python
print(classification_report(y_train, y_pred_train))
print("###"*20)
print(classification_report(y_test, y_pred_test))
```

### Detecting Optimal K value

```python
error = []

# Calculating error for K values between 1 and 40
for i in range(1, 40):  
    knn_model = KNeighborsClassifier(n_neighbors=i)
    knn_model.fit(x_train, y_train)
    y_test_pred_knn = knn_model.predict(x_test)
    error.append(np.mean(y_test_pred_knn != y_test))
```

```python
plt.figure(figsize=(12, 6))  
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',  
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate vs K Value')  
plt.xlabel('K Value')  
plt.ylabel('Mean Error')  
plt.show()
```
